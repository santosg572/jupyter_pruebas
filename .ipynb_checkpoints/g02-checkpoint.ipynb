```python
import matplotlib.pyplot as plt
import numpy as np

def Encuentra_xy(p2=0):
  x2 = p2[0]
  y2 = p2[1]
  A = (x2**2 + y2**2)/y2
  B = x2/y2
  a = 1 + B**2
  b = 2*y2*B - 2*x2 - 2*A*B
  c = x2**2 + A**2 -2*y2*A + y2**2 -1
  x1 = (-b + np.sqrt(b**2-4*a*c))/(2*a)
  x2 = (-b - np.sqrt(b**2-4*a*c))/(2*a)
  y1 = A - B*x1
  y2 = A - B*x2
  return np.round([(x1, y1), (x2, y2)],2)

def distancia(p1=0, p2=0):
  return np.round((p2[1]-p1[1])**2 + (p2[0]-p1[0])**2,2) 
  
    
def graf(p2=0, p3=0):
    x2 = p2[0]
    y2 = p2[1]
    x3 = p3[0]
    y3 = p3[1]
    plt.plot((0,x2), (0, y2))
    plt.plot((0,x3), (0, y3))
    plt.plot((x2,x3), (y2, y3))
    

res = Encuentra_xy([1,1])
print(res)

plt.plot((0,1),(0,0))
plt.plot((1,1),(0,1))
plt.plot((0,1),(0,1))

p2 = [1,1]

for i in range(5):
  print('distancia p2 a 0: ', distancia(p2, [0,0]))
  p3 = res[1]
  print('distancia p3 a 0: ',distancia(p3, [0,0]))
  print('distancia p3 a p2: ',distancia(p3, p2))
  graf(p2, p3)
  p2 = p3
  res = Encuentra_xy(p2)
```

$P_1=(0,0), P_2=(x_2, y_2), P_3=(x,y)$

Se debe cumplir: $|P_2 - P_1|^2 + |P_3 - P_2|^2 = |P_3 - P_1|$ y $|P_3  - P_2| = 1$

Entonces: $x_2^2 + y_2^2 + (x-x_2)^2 + (y-y_2)^2 = x^2 + y^2$, luego

$x_2^2 + y_2^2 - x x_2 - y y_2=0$ entonces $y = \frac{x_2^2 + y_2^2 - x x_2}{y_2}$ (1)

Sea

$$
A = \frac{x_2^2 + y_2^2}{y_2} \hspace{5mm} B = \frac{x_2}{y_2}
$$

lugo $y = A -Bx$

También se debe cumplir: $|P_3-P_2| = 1$

Luego $x^2 - 2x x_2 + x_2^2 + A^2 - 2ABx + B^2x^2 - 2y_2A + 2y_2Bx + y_2^2 = 1$

agrupando: $(1+B^2)x^2 + x (2y_2 B - 2x_2 -2AB) + (x_2^2 + A^2 - 2y_2 A + y_2^2 -1) = 0$

Resolviendo para $x$ y sustituyendo en (1) obtenemos el valor de $y$.
